```python
file_path = "../data/creditcard.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
spark_df = spark.createDataFrame(df)
spark_df.show(5)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType
import uuid

# Generate UUID as string first
from pyspark.sql.functions import udf, lit
uuid_udf = udf(lambda _: str(uuid.uuid4()), StringType())
spark_df = spark_df.withColumn("transaction_id", uuid_udf(lit(1)))

# Cast transaction_id to Cassandra UUID when writing

from pyspark.sql.functions import from_unixtime

spark_df = spark_df.withColumn("timestamp", from_unixtime(col("Time")).cast("timestamp"))
spark_df = spark_df \
    .withColumnRenamed("Amount", "amount") \
    .withColumnRenamed("is_fraud", "is_fraud")  # already correct
spark_df = spark_df.select("transaction_id", "amount", "is_fraud", "timestamp", "features")
spark_df.write \
    .format("org.apache.spark.sql.cassandra") \
    .options(table="transactions", keyspace="fraud_detection") \
    .mode("append") \
    .save()

```

```python
df_cassandra = spark.read \
    .format("org.apache.spark.sql.cassandra") \
    .options(table="transactions", keyspace="fraud_detection") \
    .load()

df_cassandra.show(5)
```
